```python
import stanza

nlp_pos = stanza.Pipeline("en", processors="tokenize,lemma,pos,ner")
filename = "../../figures/nyt_data.csv"
with open(filename, "r") as fichier:
    doc = fichier.read()


def process_text(fichier):
    """
    La fonction process_text() permet de traiter le texte d'un fichier donné en utilisant Stanza pour l'analyse syntaxique. Puis, affiche les
    mots avec leurs lemmes et leurs étiquettes de partie du discours (POS).

    Paramètres:
    fichier (str): Le chemin du fichier texte à traiter.
    """
    # Traiter le texte avec Stanza
    doc = nlp_pos(fichier)

    # Afficher les résultats
    for sentence in doc.sentences:
        for word in sentence.words:
            print(f"Mot : {word.text}\tLemme : {word.lemma}\tPOS : {word.pos}\t")


process_text(doc)
```
